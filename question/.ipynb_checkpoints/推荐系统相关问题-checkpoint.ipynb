#### 2.Exploitation & Exploration?
Exploitation： 选择现在可能的最佳方案，开发用户现有兴趣。

Exploration：选择现在不确定但未来可能高收益的方案。

在对这两类决策过程中，不断更新对所有决策的认知，优化长期目标。

#### 3.Bandit算法原理

1. 朴素bandit算法

先随机试验若干次，计算每个臂的平均收益，一直选均值最大的那个臂。

2. Epsilon Greedy（朴素的Bandit算法，有点类似模拟退火的思想）

选一个(0,1)之间较小的数epsilon，每次以epsilon的概率在所有臂中随机选一个，以1-epsilon的概率选取当前平均收益最大的臂。这里epsilon的值可以控制对exploit和explore的偏好程度，越接近0，越保守；

比如将epsilon设定为0.1，保证将10%的次数用在探索上，90%的次数用在开发上。

3. Upper Confidence Bound（置信区间上界）

初始化：先对每一个臂都试一遍按照公式计算每个臂的分数，然后选择分数最大的臂作为选择

$\overline{ X_{j}}(t)+\sqrt{\frac{2lnt}{T_{j,t}}}$

观察选择结果，更新t和T_{j,t}，其中加号前面的是这个臂到目前的收益均值，后面的叫做bonus；本质上是均值的标准差，t是目前的试验次数，$T_{j,t}$是这个臂的被试次数。这个公式反映了一个特点：均值越大，标准差越小，被选中的概率会越来越大，同时那些被选次数较少的臂也会得到试验机会。

4. Thompson Sampling

假设每个臂产生收益其背后有一个概率分布，产生收益的概率为p
不断地试验，估计出一个置信度较高的"概率p的概率分布"可以近似解决这个问题
怎么估计"概率p的概率分布"？假设概率p的概率分布符合beta(wins,lose)分布
每个臂都维护一个beta分布的参数，每次试验摇一下臂，有收益则该臂wins增加一，否则lose增加一
每次选择臂的方式是，用臂现有的beta分布产生一个随机数b，选择所有臂产生的随机数中最大的那个杆去拉

#### 4.常用的召回算法

1. memory-based CF

    1.1 user-based CF 
    
    基于用户的协同过滤，记录用户消费物品，推荐给他相似的人消费的物品；与基于物品的协同过滤比较，新颖性好，准确性差。

    2.2 item-based CF 
    
    基于物品的协同过滤，记录用户消费物品，推荐给他相似的物品。
    
    改进版I2I，要贴合业务，综合考虑 1）用户行为差 2）热门用户降权
    
    $Sim(I_{i}, I_{j}) = \frac{\sum_{u\in U_{i}\cap {U_{j}}}W_{u}^{2}/(1+σ(abs(t_{ui}-t_{uj})))}{\sqrt{\sum_u\in U(I_{i})W_{u}^{2}} \sqrt{\sum_u\in U(I_{j})W_{u}^{2}}}$
    
    $W_{u} = \frac{1}{log_{2}(3+q_{u})}$

2. Model-based CF

    从用户物品关系矩阵中学习到一个模型，从而将矩阵空白处填满。直观地说就是将原来的大矩阵，分解成两个小矩阵。
    
    2.1 MF
    
        Matrix Factorization
        2.5.1 用latent vector来表示user和item（ID embedding）
        2.5.2 组合关系用内积innter product（衡量user对于某一类商品的偏好）
    
    2.2 SVD
    
    预测打分 = 全局平均分 + 物品评分偏置 + 用户评分偏置 + 用户和物品之间的兴趣偏好

    这是增加了偏执信息的SVD，同时也加入了正则项防止过拟合。

    2.3 SVD++
    
    在SVD中结合用户的隐反馈行为和属性。

    2.4 考虑时间因素的SVD
    
    对评分按照时间加权，让久远的评分更趋近于平均值。
    
    2.5 ALS
        
        交替最小二乘矩阵分解
       
    2.6 FM
    
    因子分解机（Factorization Machine, FM），旨在解决大规模稀疏数据下的特征组合问题。
    
前面的矩阵分解算法更多地在优化均方误差，但在排序过程里，一个低MSE模型不一定能代表排序效果好。

可能原因：
    
    1. 均方误差（RMSE）和排序指标有分歧
    2. 观察有偏 - 用户总是去对喜欢的电影打分
    
故，现在的工作开始逐步朝着优化pairwise ranking loss，优化相对序关系，而不是优化绝对值。
数据规模不是很大的时候，Deep Learning可能跑出很差的结果。
